```python
#import the libaries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the file
df = pd.read_csv("ChatGPT_Reviews.csv")

#split the Review date into two 

df["Date"]=df["Review Date"].str[:9]
df["Time"]=df["Review Date"].str[9:]

# Remove the original 'Review Date' column
df.drop(columns=['Review Date'], inplace=True)

# Now remove cols with duplicates values
df= df.drop_duplicates(subset=['Date', "Time", "Review"])

df
```

```python
# Count occurrences of each rating (1 to 5)
rating_counts = df['Ratings'].value_counts().sort_index()

# Plotting the data as a bar chart
plt.bar(rating_counts.index, rating_counts.values, color='skyblue', edgecolor='black')

plt.xlabel("Ratings", fontsize= 12)
plt.ylabel("Reviews", fontsize= 12)

# Show the plot
plt.show()
```
